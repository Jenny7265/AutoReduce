# Parameter Sensitivity

This notebook uses the sensitivity solver for a one-state decay model. Sensitivity analysis is separated from the QSSA examples so that each notebook has one clear purpose.


## Model

The model is

$$
\dot{x} = -kx,
$$

with solution $x(t)=x_0e^{-kt}$. The sensitivity of the state with respect to $k$ is

$$
\frac{\partial x}{\partial k} = -t x_0 e^{-kt}.
$$


```python
import numpy as np
from sympy import Symbol

from autoreduce.system.system import System
from autoreduce.utils.reduction import get_SSM

x = Symbol("x")
k = Symbol("k")

system = System(
    [x],
    [-k * x],
    params=[k],
    params_values=[0.4],
    x_init=[5.0],
    C=np.array([[1.0]]),
)
```

## Compute Local Sensitivities

`compute_SSM` returns an array indexed by time point, parameter, and state.


```python
timepoints = np.linspace(0.0, 5.0, 6)
ssm = get_SSM(system, timepoints)
sensitivities = ssm.compute_SSM(normalize=False)

sensitivities.shape
```

## Compare Against The Analytical Sensitivity

For this model, the analytical expression is available, so it provides a compact check on the numerical sensitivity calculation.


```python
x0 = system.x_init[0]
k_value = system.params_values[0]
analytical = -timepoints * x0 * np.exp(-k_value * timepoints)
numerical = sensitivities[:, 0, 0]

np.column_stack([timepoints, numerical, analytical])
```
